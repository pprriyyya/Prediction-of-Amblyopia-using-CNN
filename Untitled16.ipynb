```python
import os
import numpy as np
import nibabel as nib
from nilearn.input_data import NiftiLabelsMasker
from nilearn import datasets

atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')
masker = NiftiLabelsMasker(atlas.maps, standardize=True)

X_all = []
y_all = []

base_dir = "/content/drive/MyDrive/Dataset1"

for label, group in enumerate(["Normal", "Amblyopia"]):
    group_path = os.path.join(base_dir, group)

    for subject in os.listdir(group_path):
        fmri_path = os.path.join(
            group_path, subject, "pre", "restingEyeOpen.nii.gz"
        )

        if os.path.exists(fmri_path):
            fmri_img = nib.load(fmri_path)
            roi_ts = masker.fit_transform(fmri_img)

            corr = np.corrcoef(roi_ts.T)
            features = corr[np.triu_indices_from(corr, k=1)]

            X_all.append(features)
            y_all.append(label)

print("Total samples:", len(X_all))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np
from collections import Counter # Import Counter for filtering

# The error 'ValueError: setting an array element with a sequence. The requested array has an inhomogeneous shape after 1 dimensions.'
# indicates that not all feature vectors in X_all have the same length.
# This means the feature extraction process in the previous cell (yBuyXpaiZFjv)
# produced feature vectors of varying lengths.

# To resolve this *within this cell*, we will filter X_all and y_all
# to ensure all feature vectors have a consistent length.
# We'll determine the most frequent length and keep only those.

if len(X_all) > 0:
    feature_lengths = [len(x) for x in X_all]
    length_counts = Counter(feature_lengths)
    most_common_length = length_counts.most_common(1)[0][0]

    # Filter X_all and y_all to only include samples with the most common feature length
    X_filtered = []
    y_filtered = []
    for i, features in enumerate(X_all):
        if len(features) == most_common_length:
            X_filtered.append(features)
            y_filtered.append(y_all[i])

    print(f"Original samples: {len(X_all)}, Filtered samples: {len(X_filtered)} (keeping feature vectors of length {most_common_length})")
    if len(X_filtered) == 0:
        raise ValueError("No samples left after filtering for consistent feature length. Cannot proceed with classification.")

    X = np.array(X_filtered)
    y = np.array(y_filtered)
else:
    raise ValueError("X_all is empty. Cannot proceed with classification.")


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

y_pred = svm.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
X_all = []
y_all = []
expected_length = None

for label, group in enumerate(["Normal", "Amblyopia"]):
    group_path = os.path.join(base_dir, group)

    for subject in os.listdir(group_path):
        fmri_path = os.path.join(
            group_path, subject, "pre", "restingEyeOpen.nii.gz"
        )

        if not os.path.exists(fmri_path):
            continue

        try:
            fmri_img = nib.load(fmri_path)
            roi_ts = masker.fit_transform(fmri_img)

            corr = np.corrcoef(roi_ts.T)

            # Replace NaNs with 0
            corr = np.nan_to_num(corr)

            features = corr[np.triu_indices_from(corr, k=1)]

            # Fix feature length consistency
            if expected_length is None:
                expected_length = len(features)

            if len(features) == expected_length:
                X_all.append(features)
                y_all.append(label)
            else:
                print(f"Skipping {subject}: feature size mismatch")

        except Exception as e:
            print(f"Skipping {subject}: {e}")

print("Final usable samples:", len(X_all))
print("Feature length:", expected_length)
```

```python
import nibabel as nib
import numpy as np
from nilearn.input_data import NiftiLabelsMasker
from nilearn import datasets
import joblib

def predict_amblyopia(fmri_path):
    # Load saved model
    svm = joblib.load("/content/drive/MyDrive/svm_amblyopia_model.pkl")
    scaler = joblib.load("/content/drive/MyDrive/scaler.pkl")

    # Load atlas
    atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')
    masker = NiftiLabelsMasker(
    atlas.maps,
    standardize='zscore_sample'
)

    # Load fMRI
    fmri_img = nib.load(fmri_path)

    # ROI extraction
    roi_ts = masker.fit_transform(fmri_img)

    # Connectivity
    corr = np.corrcoef(roi_ts.T)
    features = corr[np.triu_indices_from(corr, k=1)].reshape(1, -1)

    # Scale
    features_scaled = scaler.transform(features)

    # Predict
    pred = svm.predict(features_scaled)[0]
    prob = svm.predict_proba(features_scaled)[0][pred]

    return pred, prob
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 16, 3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(32 * 12 * 12, 64)
        self.fc2 = nn.Linear(64, 2)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        return self.fc2(x)
```

```python
# =====================================
# 0. IMPORT LIBRARIES
# =====================================
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# =====================================
# 1. CREATE FEATURES & LABELS
# (REPLACE THIS WITH YOUR REAL DATA LATER)
# =====================================
np.random.seed(42)

num_samples = 40          # total subjects
num_rois = 48             # Harvard atlas ROIs

# Fake connectivity matrices (N, 48, 48)
features = np.random.rand(num_samples, num_rois, num_rois)

# Labels: 0 = Normal, 1 = Amblyopia
labels = np.array([0]*20 + [1]*20)


# =====================================
# 2. TRAIN-TEST SPLIT
# =====================================
X_train, X_test, y_train, y_test = train_test_split(
    features, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)


# =====================================
# 3. -------- SVM BASELINE ----------
# =====================================
X_train_svm = X_train.reshape(X_train.shape[0], -1)
X_test_svm  = X_test.reshape(X_test.shape[0], -1)

scaler = StandardScaler()
X_train_svm = scaler.fit_transform(X_train_svm)
X_test_svm  = scaler.transform(X_test_svm)

svm = SVC(kernel="linear")
svm.fit(X_train_svm, y_train)

y_pred_svm = svm.predict(X_test_svm)

svm_acc  = accuracy_score(y_test, y_pred_svm)
svm_prec = precision_score(y_test, y_pred_svm)
svm_rec  = recall_score(y_test, y_pred_svm)
svm_f1   = f1_score(y_test, y_pred_svm)


# =====================================
# 4. PREPARE DATA FOR CNN / RESNET
# =====================================
X_train_tensor = torch.tensor(X_train, dtype=torch.float32).unsqueeze(1)
X_test_tensor  = torch.tensor(X_test, dtype=torch.float32).unsqueeze(1)

y_train_tensor = torch.tensor(y_train, dtype=torch.long)
y_test_tensor  = torch.tensor(y_test, dtype=torch.long)


# =====================================
# 5. -------- CNN MODEL ----------
# =====================================
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 16, 3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(32 * 12 * 12, 64)
        self.fc2 = nn.Linear(64, 2)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        return self.fc2(x)


model_cnn = SimpleCNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_cnn.parameters(), lr=0.001)

for epoch in range(15):
    optimizer.zero_grad()
    outputs = model_cnn(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()


with torch.no_grad():
    outputs = model_cnn(X_test_tensor)
    _, y_pred_cnn = torch.max(outputs, 1)

cnn_acc  = accuracy_score(y_test, y_pred_cnn.numpy())
cnn_prec = precision_score(y_test, y_pred_cnn.numpy())
cnn_rec  = recall_score(y_test, y_pred_cnn.numpy())
cnn_f1   = f1_score(y_test, y_pred_cnn.numpy())


# =====================================
# 6. -------- RESNET MODEL ----------
# =====================================
class ResidualBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.conv1 = nn.Conv2d(channels, channels, 3, padding=1)
        self.conv2 = nn.Conv2d(channels, channels, 3, padding=1)

    def forward(self, x):
        identity = x
        x = torch.relu(self.conv1(x))
        x = self.conv2(x)
        x += identity
        return torch.relu(x)


class SmallResNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = nn.Conv2d(1, 32, 3, padding=1)
        self.res1 = ResidualBlock(32)
        self.res2 = ResidualBlock(32)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc = nn.Linear(32 * 24 * 24, 2)

    def forward(self, x):
        x = torch.relu(self.conv(x))
        x = self.res1(x)
        x = self.res2(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1)
        return self.fc(x)


model_resnet = SmallResNet()
optimizer = optim.Adam(model_resnet.parameters(), lr=0.001)

for epoch in range(15):
    optimizer.zero_grad()
    outputs = model_resnet(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()


with torch.no_grad():
    outputs = model_resnet(X_test_tensor)
    _, y_pred_resnet = torch.max(outputs, 1)

res_acc  = accuracy_score(y_test, y_pred_resnet.numpy())
res_prec = precision_score(y_test, y_pred_resnet.numpy())
res_rec  = recall_score(y_test, y_pred_resnet.numpy())
res_f1   = f1_score(y_test, y_pred_resnet.numpy())


# =====================================
# 7. RESULTS
# =====================================
print("\nMODEL COMPARISON\n")

print("SVM   -> Acc:", svm_acc, "Prec:", svm_prec, "Rec:", svm_rec, "F1:", svm_f1)
print("CNN   -> Acc:", cnn_acc, "Prec:", cnn_prec, "Rec:", cnn_rec, "F1:", cnn_f1)
print("ResNet-> Acc:", res_acc, "Prec:", res_prec, "Rec:", res_rec, "F1:", res_f1)
```

```python
def predict_amblyopia(connectivity_matrix):
    """
    Input:
        connectivity_matrix -> numpy array of shape (48, 48)

    Output:
        Dictionary with predictions from SVM, CNN, ResNet
        + Final majority-vote decision
    """

    results = {}

    # ---------- SVM ----------
    x_svm = connectivity_matrix.reshape(1, -1)
    x_svm = scaler.transform(x_svm)
    svm_pred = svm.predict(x_svm)[0]
    results["SVM"] = "Amblyopia" if svm_pred == 1 else "Normal"

    # ---------- CNN ----------
    model_cnn.eval()
    x_cnn = torch.tensor(connectivity_matrix, dtype=torch.float32)
    x_cnn = x_cnn.unsqueeze(0).unsqueeze(0)

    with torch.no_grad():
        out_cnn = model_cnn(x_cnn)
        _, cnn_pred = torch.max(out_cnn, 1)

    results["CNN"] = "Amblyopia" if cnn_pred.item() == 1 else "Normal"

    # ---------- ResNet ----------
    model_resnet.eval()
    x_res = torch.tensor(connectivity_matrix, dtype=torch.float32)
    x_res = x_res.unsqueeze(0).unsqueeze(0)

    with torch.no_grad():
        out_res = model_resnet(x_res)
        _, res_pred = torch.max(out_res, 1)

    results["ResNet"] = "Amblyopia" if res_pred.item() == 1 else "Normal"

    # ---------- FINAL DECISION (Majority Vote) ----------
    votes = [svm_pred, cnn_pred.item(), res_pred.item()]
    final_pred = 1 if votes.count(1) >= 2 else 0

    results["Final Decision"] = "Amblyopia" if final_pred == 1 else "Normal"

    return results
```
